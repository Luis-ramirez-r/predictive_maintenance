```python
def perform_pca(df, feature_cols, response_cols, num_components=2):
    # Estandarizar las variables de características
    scaler = StandardScaler()
    df_norm = scaler.fit_transform(df[feature_cols])

    # Realizar el PCA con el número especificado de componentes principales
    pca = PCA(n_components=num_components)
    pca.fit(df_norm)
    df_pca = pca.transform(df_norm)
    pca_cols = ['PC' + str(i+1) for i in range(num_components)]
    df_pca = pd.DataFrame(df_pca, columns=pca_cols)

    # Agregar las variables de respuesta
    for col in response_cols:
        df_pca[col] = df[col].values

    return df_pca
```

```python
response_cols = ['failure_binary', 'failure']
df_dataset_telemetria_pca = perform_pca(df_dataset_telemetria, base_fetures, response_cols)
```

```python
features_agg= ['mean_volt', 'mean_rot', 'mean_press', 'mean_vib', 'max_volt',
       'max_rot', 'max_press', 'max_vib', 'min_volt', 'min_rot', 'min_press',
       'min_vib', 'std_volt', 'std_rot', 'std_press', 'std_vib']

datos_telemetria_agg_pca = perform_pca(datos_telemetria_agg, features_agg, response_cols)
```
