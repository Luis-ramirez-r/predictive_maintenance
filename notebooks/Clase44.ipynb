```python
def plot_pca_clusters(n_clusters):
    # split X and y
    X = df_failures_telemetry_agg.drop(columns=['failure', 'failure_binary'])
    y_bin = df_failures_telemetry_agg['failure_binary']
    # apply StandardScaler
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    # perform PCA
    pca = PCA()
    X_pca = pca.fit_transform(X_scaled)
    # convert the transformed data to a DataFrame
    columns = [f'PC{i+1}' for i in range(X_pca.shape[1])]
    df_pca = pd.DataFrame(X_pca, columns=columns)
    # train a clustering model with specified number of clusters on PCA data
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X_pca)
    cluster_labels = kmeans.labels_
    centroids = kmeans.cluster_centers_
    # add cluster labels to the PCA DataFrame
    df_pca['Cluster'] = cluster_labels
    # plot PCA1 vs PCA2 and use color to show assigned cluster
    scatter = plt.scatter(df_pca['PC1'], df_pca['PC2'], c=cluster_labels, cmap='viridis')
    # plot the centroids
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', label='Centroids')
    # add axis labels, color bar, legend, and title
    plt.xlabel('PCA1')
    plt.ylabel('PCA2')
    plt.colorbar(scatter, label='Assigned Cluster')
    plt.legend()
    plt.title(f'PCA1 vs PCA2 - KMeans Clustering with {n_clusters} Clusters (PCA Data)')
    # display the plot
    plt.show()
    # return the DataFrame with PCA data and cluster labels
    return df_pca
# Call the function with desired number of clusters and store the resulting DataFrame
df_pca_clusters = plot_pca_clusters(5)

```

# Outliers 

```python
from sklearn.ensemble import IsolationForest
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Generate some example data
X = np.random.randn(1000, 2)  # 1000 samples, 2 features

# Create and train the model
model = IsolationForest(n_estimators=100, contamination=0.002)
model.fit(X)

# Predict the outlier score for each sample
scores = model.score_samples(X)

# Plot the outlier scores as a scatter plot
plt.scatter(X[:, 0], X[:, 1], c=scores, cmap='coolwarm')
plt.colorbar()
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Isolation Forest Outlier Scores')
plt.show()
```
