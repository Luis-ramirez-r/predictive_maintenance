```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

```

```python
# Global options.
plt.rcParams['font.family'] = 'serif'

# Data from p74 of Visual Display of Quantitative Information.
x = list(range(1951, 1960))
y = [264, 231, 274, 241, 322, 286, 283, 247, 243]

# Plot line, line masks, then dots.
fig, ax = plt.subplots()
ax.plot(x, y, linestyle='-', color='black', linewidth=1, zorder=1)
ax.scatter(x, y, color='white', s=100, zorder=2)
ax.scatter(x, y, color='black', s=20, zorder=3)

# Remove axis lines.
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Set spine extent.
ax.spines['bottom'].set_bounds(min(x), max(x))
ax.spines['left'].set_bounds(225, 325)

# Reduce tick spacing.
x_ticks = list(range(min(x), max(x)+1, 2))
ax.xaxis.set_ticks(x_ticks)
ax.yaxis.set_major_locator(ticker.MultipleLocator(base=25))
ax.tick_params(direction='in')

# Adjust lower limits to let data breathe.
ax.set_xlim([1950, ax.get_xlim()[1]])
ax.set_ylim([210, ax.get_ylim()[1]])

# Axis labels as a title annotation.
ax.text(1958, 320, 'Connecticut Traffic Deaths,\n1951 - 1959')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

all_precip_data = pd.read_csv("https://johndjohnson.info/files/MilwaukeePrecipitationData2020.csv")

legend_data = all_precip_data.loc[all_precip_data['month'] == 3].copy()
legend_data['dayofleapyear'] = legend_data['dayofleapyear'] - 15
legend_data['normal_precip'] = legend_data['normal_precip'] + 30
legend_data['max_cum_precip'] = legend_data['max_cum_precip'] + 30
legend_data['low_cum_precip'] = legend_data['low_cum_precip'] + 30
legend_data['cum_precip_2020'] = legend_data['cum_precip_2020'] + 30

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(legend_data['dayofleapyear'], legend_data['normal_precip'], color='gray40', linewidth=1.5)
ax.plot(legend_data['dayofleapyear'], legend_data['cum_precip_2020'], linewidth=1)
ax.fill_between(legend_data['dayofleapyear'], legend_data['low_cum_precip'], 
                legend_data['max_cum_precip'], alpha=0.25)
ax.fill_between(legend_data['dayofleapyear'], legend_data['normal_precip'] - legend_data['ytd_precip_sd'], 
                legend_data['normal_precip'] + legend_data['ytd_precip_sd'], alpha=0.5)

plt.show()
```
